```python
import pandas as pd
from glob import glob
```

```python
data_dict = pd.read_csv('../data/raw_pems/meta_dict.csv', names=['Feature', 'Description']).iloc[1:].reset_index(drop=True)
data_dict
```

```python
meta_txt_files = glob('../data/raw_pems/*.txt')
meta_txt_files
```

```python
meta = pd.read_csv(meta_txt_files[0], delimiter='\t')
meta.info()
```

```python
meta.head()
```

```python
meta = meta.drop(columns=['User_ID_2', 'User_ID_3', 'User_ID_4'])
meta = meta[(~meta['Longitude'].isna()) & (~meta['Latitude'].isna())].reset_index(drop=True)
```

```python
import pickle
```

```python
with open('../data/loaded/meta.dat', 'wb') as f:
    pickle.dump(meta[['ID', 'Fwy', 'Dir', 'Latitude', 'Longitude', 'Length', 'Type', 'Lanes', 'Abs_PM']], f)
```
